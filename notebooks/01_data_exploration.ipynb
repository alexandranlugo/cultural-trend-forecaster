```python
import pandas as pd
import os

os.chdir(os.path.dirname(os.getcwd()))

lastfm  = pd.read_csv("data/raw/lastfm_artists.csv")
youtube = pd.read_csv("data/raw/youtube_videos.csv")
news    = pd.read_csv("data/raw/news_headlines.csv")

print("LASTFM")
print(lastfm[["artist_name","listeners","playcount","plays_per_listener"]].sort_values("plays_per_listener", ascending=False).to_string(index=False))

print("\nYOUTUBE: total views per artist")
print(youtube.groupby("artist_name")["view_count"].sum().sort_values(ascending=False).to_string())

print("\nYOUTUBE: avg views per day (velocity)")
print(youtube.groupby("artist_name")["views_per_day"].mean().sort_values(ascending=False).to_string())

print("\nNEWS: headline count per artist")
print(news.groupby("artist_name").size().sort_values(ascending=False).to_string())
```

* Mk.gee is the most interesting in the whole dataset: Lowest listener count on the list at 726K but 26M YouTube views and 46 plays per listener -- That's a cult artist on the verge of breaking wide
* Sabrina Carpenter dominates every single metric: highest plays per listener, highest YouTube views, most Last.fm listeners. Gives model a clear "positive" example to learn from because that's what a fully broken-out artist looks like in the data.
* Ethel Cain and Gracie Abrams have high plays per listener but relatively modest YouTube numbers, suggesting a dedicated streaming fanbase that hasn't crossed over to casual video viewers yet.
